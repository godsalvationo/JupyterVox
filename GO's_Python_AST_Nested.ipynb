```python
import ast

# generate dict key for statements
def gen_Dict_Key(node, level):
  if isinstance(node, ast.Expr):
    return "ExprStmt-" + str(level)
  elif isinstance(node, ast.BinOp):
    return "BinOpStmt-" + str(level)

# parse an ast.Add/Mult/... node
def parse_Opcode(node):
  if isinstance(node, ast.Add):
    return 'plus'
  elif isinstance(node, ast.Mult):
    return 'multiply'
  elif isinstance(node, ast.Sub):
    return "subtract"
  elif isinstance(node, ast.Div):
    return "divide"
  else:
    raise Exception("Unknown Opcode" + str(node))


# parse an ast.Name code
def parse_Name(node):
  return node.id

# parse an binary operation
def parse_BinOp(node, level):
 # create an empty array to hold the values
  speech = {}

  # BinOp has left, op, and righ fileds
  # visit left
  if isinstance(node.left, ast.Name):
    speech['left'] = parse_Name(node.left)
  elif isinstance(node.left, ast.BinOp):
    speech['left'] = ast_walk(node.left, level+1)

  # visit operation
  speech['op'] = parse_Opcode(node.op)

  # visit right
  if isinstance(node.right, ast.Name):
    speech['right'] = parse_Name(node.right)
  elif isinstance(node.right, ast.BinOp):
    speech['right'] = ast_walk(node.right, level+1)

  return speech

# tree walk
def ast_walk(node, level=0):
  # create an empty array to hold the values
  speech = {}

  if isinstance(node, ast.BinOp):
    speech["BinOp-" + str(level)] = parse_BinOp(node, level=level)
    #print(speech)

    return speech

  # walk the tree
  for field, value in ast.iter_fields(node):
    if isinstance(value, list):
      for item in value:
        if isinstance(item, ast.AST):
          speech[gen_Dict_Key(item, level)] = ast_walk(item, level=level+1)
    elif isinstance(value, ast.AST):
      # start handling different statements
      speech[gen_Dict_Key(value, level)] = ast_walk(value, level=level+1)

  #print(speech)
  return speech

speech = ast_walk(ast.parse('a*b-c'))

print(speech)
```
